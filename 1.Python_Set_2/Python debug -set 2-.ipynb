```python
#2.Find the error in the following Python code:
'''def my_function():
    x += 5
    print(x)

x = 10
my_function()'''

# x is not declared inside the function class, So x declared as global issue fixed.

def my_function():
    global x
    x += 5
    print(x)

x = 10
my_function()
```
